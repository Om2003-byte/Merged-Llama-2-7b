```python
import json
import os

def create_config_json(path):
    config_path = os.path.join(path, "config.json")
    if not os.path.exists(config_path):
        config = {
            "architectures": ["LlamaForCausalLM"],
            "model_type": "llama",
            "torch_dtype": "float16"
        }
        with open(config_path, "w") as f:
            json.dump(config, f, indent=2)
        print(f"✅ Created config.json at {config_path}")
    else:
        print(f"ℹ️ config.json already exists at {config_path}")

create_config_json("/content/llama-2-7b")
create_config_json("/content/llama-2-7b-chat")
```

```python
config_content = """
base_model: meta-llama/Llama-2-7b-hf
merge_method: slerp
dtype: float16

models:
  - model: meta-llama/Llama-2-7b-hf
    weight: 0.5
  - model: meta-llama/Llama-2-7b-chat-hf
    weight: 0.5

parameters:
  t: 0.5
"""

with open("/content/linear2.yml", "w") as f:
    f.write(config_content)
```

```python
import torch
import yaml
from mergekit.config import MergeConfiguration
from mergekit.merge import MergeOptions, run_merge

CONFIG_YML = "/content/linear2.yml"

with open(CONFIG_YML, "r", encoding="utf-8") as fp:
    merge_config = MergeConfiguration.model_validate(yaml.safe_load(fp))

run_merge(
    merge_config,
    out_path="./models/merged-llama-2-7b",
    options=MergeOptions(
        lora_merge_cache="/tmp",
        cuda=torch.cuda.is_available(),
        copy_tokenizer=True,
        lazy_unpickle=False,
        low_cpu_memory=False,
    ),
)

print("✅ Merge complete!")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_path = "./models/merged-llama-2-7b"
model = AutoModelForCausalLM.from_pretrained(model_path, torch_dtype=torch.float16, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_path)

prompt = "### Instruction:\nExplain how photosynthesis works in simple terms.\n\n### Response:\n"
inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

outputs = model.generate(**inputs, max_new_tokens=200, do_sample=True, temperature=0.7)
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
